```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.metrics import confusion_matrix
import seaborn as sns
```

```python
validation_results_path = '/home/b10173209/test.file/X/validation_results_lr_0.1_md_9_ne_300.xlsx'
data = pd.read_excel(validation_results_path)
data.head()
```

```python
data['Emotion'] = data['Folder'].str.split('_').str[0]
emotions = data['Emotion'].unique() 
emotion_dataframes = {emotion: data[data['Emotion'] == emotion] for emotion in emotions} 
print(data['Emotion'].head())
```

```python
for emotion, df in emotion_dataframes.items():
    # 提取真實標籤和預測標籤
    true_labels = df['Labels']
    predicted_labels = df['Predictions']
    
    # 生成混淆矩陣
    cm = confusion_matrix(true_labels, predicted_labels, labels=true_labels.unique())
    
    # 繪製混淆矩陣
    plt.figure(figsize=(8, 6))
    sns.heatmap(cm, annot=True, fmt='d', cmap='Greys',
                xticklabels=true_labels.unique(), yticklabels=true_labels.unique())
    plt.title(f'Confusion Matrix for {emotion}')
    plt.xlabel('Predicted Labels')
    plt.ylabel('True Labels')
    plt.savefig(f'{emotion}_XGBoost_Confusion Matrix.png', dpi=300, bbox_inches='tight')
    plt.show()
```

```python
label_mapping = {
    0: 'anger',
    1: 'fear',
    2: 'joy',
    3: 'love',
    4: 'sadness',
    5: 'surprise'
}
```

```python
data['Labels'] = data['Labels'].map(label_mapping)
data['Predictions'] = data['Predictions'].map(label_mapping)

# Generate confusion matrix
unique_labels = list(label_mapping.values())
cm = confusion_matrix(data['Labels'], data['Predictions'], labels=unique_labels)
```
