```python
import pandas as pd
import matplotlib.pyplot as plt
from matplotlib import rcParams
from matplotlib.font_manager import FontProperties
```

```python
# 讀取資料
train = pd.read_csv('/home/b10173209/test.file/train.txt', sep=';', names=['text', 'label'])
test = pd.read_csv('/home/b10173209/test.file/test.txt', sep=';', names=['text', 'label'])
val = pd.read_csv('/home/b10173209/test.file/val.txt', sep=';', names=['text', 'label'])

# 合併資料
data = pd.concat([train, test, val], axis=0)
```

```python
# 計算分布
label_distribution = data['label'].value_counts(normalize=True) * 100
formatted_distribution = label_distribution.apply(lambda x: f"{x:.2f}%")
print(formatted_distribution)

# 繪製圖表
plt.figure(figsize=(10, 6))
label_distribution.plot(kind='bar', color='grey')
plt.title('Emotion Label Distribution')
plt.xlabel('Emotion')
plt.ylabel('Percentage (%)')
plt.xticks(rotation=45)
plt.savefig('distribution.png', dpi=300, bbox_inches='tight')
plt.show()
```
